```python
# Line plot
# whenever time component is involved on x axis
# depict the relation between two variables in the trend

x = [1,2,3,4,5]
y = [6,1,7,8,2]

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red")
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize = 10)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
plt.plot(x,y, color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize = 10)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]

plt.plot(x,y1,color = "red", marker = "o", linestyle = '--', linewidth = 3, markersize = 6, label = 'line y1')
plt.plot(x,y2,color = "blue", marker = "o", linestyle = '--', linewidth = 3, markersize = 6, label = 'line y2')


plt.xlabel("x data axis")
plt.ylabel("y data axis")

plt.title("Multiple lines in one plot")
plt.legend() # to show the labels

plt.show()
```

```python
# bar plot >>> for categorical data, not continuous

x = ['a','b','c','d']
y = [4, 6, 1, 2]

plt.bar(x,y,color = 'r')
```

```python
# bar plot >>> for categorical data, not continuous

x = ['a','b','c','d']
y = [4, 6, 1, 2]

plt.barh(x,y,color = 'r') # use barh for horizontal plotting.
```

```python
# pie chart >>> to understand the percentage distribution of the data

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs)
plt.title ("Pie chart")
plt.show()
```

```python
# pie chart >>> to understand the percentage distribution of the data

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title ("Pie chart")
plt.show()
```

```python

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title ("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.3)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%' , explode = explode, shadow = True)
plt.title ("Pie chart")
plt.show()
```

```python
# 3d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c = 'red')
```
